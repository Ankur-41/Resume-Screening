* Overview: This dataset contains 1,000+ synthetic resumes with key details such as skills, experience, education, job roles, certifications, AI screening scores, and recruiter decisions.

* Features:
Resume_ID: Unique identifier Name: Candidate's name Skills: List of relevant technical skills Experience (Years): Total work experience Education: Highest qualification Certifications: Relevant industry certifications Job Role: Target job position Recruiter Decision: Hire or Reject Salary Expectation ($): Expected salary Projects Count: Number of projects completed AI Score (0-100): AI-based resume ranking score.

```python
def bar_plot(df,col:str,plot_no:int): 
    plt.subplot(1,2,plot_no)
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('counts')
    
```

```python
df['Certifications'] = df['Certifications'].fillna('')
```

```python
df.isnull().sum()
```

```python
df['combined'] = (
    'Skills : ' + df['Skills'] + '. '
    'Education : ' + df['Education'] + '. '
    'Experience: ' + df['Experience (Years)'].astype(str) + 'years. '
    'Certifications: ' + df['Certifications'] + '. '
    'Projects: ' + df['Projects Count'].astype(str) + 'projects. '
)
```

```python
df['combined'][0]
```

```python
import re 
import emoji
def filter_txt(txt):
    url_pattern = r'(https?://\S+|www\.\S+)|((http|https)\s*:\s*//\s*\S+)'
    txt = txt.lower()
    txt = emoji.replace_emoji(txt,'')
    txt = re.sub(r'<.*?>', ' ', txt)
    txt = re.sub(url_pattern,'',txt)
    txt = txt.replace('\n',' ')
    txt = txt.strip()
    return txt
```

```python
df['combined'] = df['combined'].apply(filter_txt)
```

```python
# pip install sentence-transformers
```

## Using Transformer Model

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('paraphrase-MPNet-base-v2')
```

```python
embedding = model.encode(
    df['combined'].tolist(),
    batch_size=32, 
    convert_to_numpy=True, 
    normalize_embeddings=True, 
    show_progress_bar=True
)
```

## Trying with an random example

```python
job_desc = '''We are looking for a Data Scientist to join our analytics team. 
Responsibilities include:

- Analyze large datasets to identify trends and actionable insights.
- Build and deploy predictive models using Python and machine learning libraries (scikit-learn, TensorFlow, Keras).
- Work with SQL and NoSQL databases to extract and manipulate data.
- Perform data preprocessing, feature engineering, and visualization.
- Collaborate with cross-functional teams to understand business requirements and deliver data-driven solutions.

Required Skills:

- Proficiency in Python and data science libraries (NumPy, Pandas, Matplotlib, Seaborn).
- Experience with machine learning models including regression, classification, clustering.
- Knowledge of deep learning frameworks like TensorFlow or Keras.
- Strong problem-solving skills and ability to communicate insights clearly.
- Familiarity with cloud platforms (AWS, GCP, or Azure) is a plus.

Education & Experience:

- Bachelor's or Master's degree in Computer Science, Statistics, Mathematics, or related field.
- 13 years of experience in data analysis or machine learning projects.

Certifications (Optional):

- Relevant certifications such as Data Science Professional, Machine Learning, or AI certifications are a plus.
'''
```

```python
cleaned_job_desc = filter_txt(job_desc)
```

```python
job_emb = model.encode(cleaned_job_desc,convert_to_numpy=True,normalize_embeddings=True,show_progress_bar=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_scores = cosine_similarity([job_emb],embedding)
```

```python
top_scores = np.argsort(similarity_scores[0])[::-1][:10]
```

```python
df.iloc[top_scores]
```
